*Import Libraries*

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
from scipy.fftpack import fft
```

```python
fs = 500  # Sampling frequency
t = np.linspace(0, 1, fs, endpoint=False)  # Time vector
egg_signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 20 * t)  # Generate an egg-shaped signal

frequencies, psd = welch(egg_signal, fs, nperseg=256, window='hann', nfft=512)# Welch's method for PSD estimation

fft_frequencies = np.fft.fftfreq(len(egg_signal), 1/fs)                       # FFT for comparison        
fft_signal = fft(egg_signal)
fft_magnitude = np.abs(fft_signal)[:len(fft_signal)//2]

# Plotting
plt.figure(figsize=(6, 4))

plt.subplot(2, 1, 1)    # Time-domain plot
plt.plot(t, egg_signal)
plt.title('Egg-shaped Signal in Time Domain', fontsize=10)
plt.xlabel('Time (s)', fontsize=10)
plt.ylabel('Amplitude', fontsize=10)

plt.subplot(2, 1, 2)   # Frequency-domain plot
plt.semilogy(frequencies, psd, label="Welch's Method")
plt.semilogy(fft_frequencies[:len(fft_frequencies)//2], fft_magnitude, label='FFT')
plt.title('Power Spectral Density (PSD)', fontsize=10)
plt.xlabel('Frequency (Hz)', fontsize=10)
plt.ylabel('PSD', fontsize=10)
plt.legend()
plt.autoscale(axis="x",tight=True, enable=True)
plt.tight_layout()
```
